```python
#import libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#load the dataset
data = pd.read_csv("payments.csv")
```

```python
data
```

```python
#remove rows that contain null values
data = data.dropna()
```

```python
data
```

```python
#use Label Encoding to convert string to numeric value
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
#replace type column with encoded values
data['type'] = label.fit_transform(data['type'])
#replace nameOrig column with encoded values
data['nameOrig'] = label.fit_transform(data['nameOrig'])
#replace nameDest column with encoded values
data['nameDest'] = label.fit_transform(data['nameDest'])
```

```python
data
```

```python
#split the dataset into training and testing sets
#take all values of columns except the last column
x = data.iloc[:, :-1].values
#take the values of the last column
y = data.iloc[:, -1].values
#split the dataset, test_size = 0.2 means 20% of the dataset is taken as test data and the balance 80% for training
#random_state determines the random seed used for shuffling data and feature selection during model training
#Set a fixed value for random_state so that it produces the same value it's run, here it runs 50 times
#if the random_state is not set it will use a different random_seed everytime it's run resulting in a variability of results
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 50)
```

```python
#scale the features
#scaling the variables brings all values to the same scale(min-max normalization)
st_sc = StandardScaler()
x_train = st_sc.fit_transform(x_train)
x_test = st_sc.transform(x_test)
```

```python
x_test
```

```python
#train the random forest classifier
rfc = RandomForestClassifier(n_estimators = 100, criterion='entropy', random_state = 50)
rfc.fit(x_train, y_train)
```

```python
#make predictions on the testing set
y_pred = rfc.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
