# Modules

```python

import  numpy as np #used to perform mathematical operations on arrays
import pandas as pd #used for data manipulation and analysis
from pydataset import data #python package that provides various open-source datasets
from sklearn import svm #svm is a supervised machine learning algorithm
from sklearn.metrics import classification_report #performance evaluation metric in machine learning
from sklearn import model_selection #method for setting a blue print to analyze data  and then using it to measure new data
```

# Data Preparation

```python
df=pd.read_csv('PS_20174392719_1491204439457_log.csv') #takes a path to the csv file and reads the data into a pandas dataframe object
df=df.dropna() #remove rows that contain null values
df.head() #returns the first 5 rows
```

# Dummy Variables

```python
dummy_df = pd.get_dummies( df,columns=['type']) #convert the 'type' catagorical column to a numerical column
```

```python
dummy_df.head #dataset with the converted dummy variables
```

```python
df=pd.concat([df,dummy_df[['type_CASH_IN','type_CASH_OUT','type_DEBIT','type_PAYMENT','type_TRANSFER']]],axis=1)
#concatenate the original dataset with the newly created 5 columns
```

```python
df.head #prints first 5 rows
```

```python
df=df.drop(['nameOrig','nameDest','type'],axis=1) #drop the catagorical columns
```

```python
df.head #print first 5 rows
```

# Scaling of Variables

```python
 #brings all values to the same scale(min-max normalization)
df=(df-df.min())/(df.max()-df.min())
```

```python

df.head() #print first 5 rows
```

# Model Development

# Independent and Dependent Variables


```python
X=df[['step','amount','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest','type_CASH_IN','type_CASH_OUT','type_DEBIT','type_PAYMENT','type_TRANSFER']]#independent variables
y=df['isFraud'] #dependent variable
```

# Train and Test Sets

```python
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=.3,random_state=1)
#split the dataset into training and testing datasets
```

# Models

```python
h1=svm.LinearSVC(C=1) #create an instance of the svm classifiere
```

```python
h1.fit(X_train,y_train) #fit the model
h1.score(X_train,y_train) #get the accuracy score
```

# Model Testing

```python
y_pred=h1.predict(X_test) #make prediction using the test data
```

```python
pd.crosstab(y_test,y_pred) #see the strength of the model
```

```python
print(classification_report(y_test,y_pred))#generate a report using the various classification metrics
```
